```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def morphShow(fname):
    file = open(fname, 'r')
    data = file.readlines()
    
    while data[0][0] == '#' or data[0][0] == '\n':
        del data[0]
            
    for i in range(0, len(data) - 1):
        node = list(map(float, data[i].split()))
        if node[0] != 1:
            parent_node = list(map(float, data[int(node[6]) - 1].split()))
            
        if node[0] == 1:
            soma = plt.Circle((node[2], node[3]), radius=node[5], fc='y')
            plt.gca().add_patch(soma)
        elif node[1] == 7:
            #if parent_node[0] == 1:
            #    R = node[5]
            #else:
            #    R = (node[5] + parent_node[5]) / 2
            segment = plt.Line2D((node[2], parent_node[2]), (node[3], parent_node[3]),
                                 lw=2*node[5])
            plt.gca().add_line(segment)

    plt.axis('equal')       
    plt.show()
    file.close()
```

```python
morphShow('A1-CA1-L-A63x1zACR3.CNG.swc.txt')
```
